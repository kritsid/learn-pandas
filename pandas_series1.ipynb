pandas Series

```
import pandas as pd
import numpy as np
```

syntax:
pd.Series(data = data,index= index,dtype = , copy=)

```
# examples:
series_from_list = pd.Series([1,2,3,4,5])
series_from_array = pd.Series(np.array([1,2,3,4,5]))
```

```
series_from_list 
series_from_array
```

```
index = ['a','b','c','d','e']
series1 = pd.Series(data = [2,4,6,8,10],index=index)
```

```
series1
```

from dictionary

When the index is not specified, then the keys are taken in a sorted order as index values.


```
dict = {'key1':'val1','key2':'val2','key3':6.7}
dict
```

```
series_from_dict = pd.Series(data=dict)
series_from_dict
```

```
index = ['key1','key3','key2','key4']
series2 = pd.Series(data=dict,index=index)
series2
```

If the index is passed, values in data corresponding to the labels in the index will be accessed, the index which is absent in the keys of the dictionary will have NaN values.


You can store heterogeneous data while creating a series with a dictionary


```
t_dict = {'a' : [1,2,3], 'b': [4,5], 'c':6, 'd': "Hello World"}
```

```
# list of lables
lables = ['a','b','c']

# dictionary
dic = {'b':1,'c':2,'d':3}

# Series without specified labels
print(pd.Series(dic))
print('============')

# Series with specified labels
print(pd.Series(dic, labels))
```

HELPER FUNCTIONS:

if we are using numpy to create a series, only homogenous data can be added 
but if we are creating it via dictionary then hetrogenous data can be added.
